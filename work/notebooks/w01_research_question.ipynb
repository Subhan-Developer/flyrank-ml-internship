## 1. My Lane

**I choose Lane 1: Content Decline Prediction**

**Why:**
- The data has a clear target: pages that are declining (`trend_direction == "down"`)
- There's already a baseline model (Precision@50 = 0.240) that I can try to beat
- This is a real business problem: finding declining content early saves money
- I have enough data (30,000 pages) to train a model

**Problem type:** Binary classification (predict decline = yes/no)

---

## 2. My Research Question

**The question:** Which pages will show declining engagement in the next period?

**What I'm analyzing:** Each individual page in the dataset

**What my model will output:** 
- A prediction: "declining" or "not declining"
- A probability score (0 to 1) to rank pages

**What someone will DO with this:**
- Content editors will take the top 50 predicted declining pages
- They will refresh/update these pages to prevent traffic loss

**What happens if I'm wrong:**
- **False alarm** (predict decline but it doesn't): Editors waste time on unnecessary refreshes
- **Missed decline** (don't predict it but it happens): Page loses traffic, revenue drops

**Why I need ML:**
- A simple rule can only catch obvious patterns (Precision@50 = 0.240)
- ML can find complex patterns and give a 3x improvement (Precision@50 = ~0.740)

```python
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv('data/raw/content_refresh_anonymized.csv')

# Number 1: How much data do we have?
print(f"📊 Total pages in dataset: {len(df):,}")

# Number 2: What percentage of pages are declining?
decline_pct = (df['trend_direction'] == 'down').mean() * 100
print(f"📉 Pages that are declining: {decline_pct:.1f}%")

# Number 3: How many features can we use?
numeric_features = len(df.select_dtypes(include=[np.number]).columns)
print(f"🔢 Number of numeric features available: {numeric_features}")

# Bonus: Show column names
print(f"\n📋 Sample of columns: {df.columns[:5].tolist()}")
```
